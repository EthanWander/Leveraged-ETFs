## Modified Geometric Annual Growth

For a standard deviation of n from the median, we have that the stock price is

$$
S_{t,n} = S_0 \cdot e^{\mu t - \frac{1}{2}\sigma^2 t + n\sigma\sqrt{t}}
$$

The number we should be considering as the annual geometric growth is then

$$
\mu_n = \frac{1}{t} \ln{\left(\frac{S_{t,n}}{S_0}\right)} = \mu - \frac{1}{2}\sigma^2 + \frac{n\sigma}{\sqrt{t}}
$$
$$
S_{t,n} = e^{\mu_n t}
$$

instead of $\mu$.

### Calculation of $\mu_{-1}$ for $t=20$

Using the formula:
$$
\mu_n = \mu - \frac{1}{2}\sigma^2 + \frac{n\sigma}{\sqrt{t}}
$$

For $n = -1$ and $t = 20$, we calculate $\mu_{-1}$ for S&P 500 and QQQ:

#### S&P 500:
- $\mu = 15.7\% = 0.157$
- $\sigma = 15\% = 0.15$
- $t = 20$

Substitute into the formula:
$$
\mu_{-1} = 0.157 - \frac{1}{2}(0.15)^2 + \frac{-1 \cdot 0.15}{\sqrt{20}}
$$
$$
\mu_{-1} = 0.157 - 0.01125 - 0.03354 \approx 0.11221 \text{ or } 11.221\%
$$
#### QQQ:
- $\mu = 22.0\% = 0.22$
- $\sigma = 22\% = 0.22$
- $t = 20$

Substitute into the formula:
$$
\mu_{-1} = 0.22 - \frac{1}{2}(0.22)^2 + \frac{-1 \cdot 0.22}{\sqrt{20}}
$$
$$
\mu_{-1} = 0.22 - 0.0242 - 0.0492 \approx 0.1466 \text{ or } 14.66\%
$$

These calculations provide the adjusted geometric growth rates $\mu_{-1}$ for both indices.

## Maximal Leverage

For a leveraged stock this would be:
$$ \mu_n = \beta\mu - \frac{1}{2}\beta^2\sigma^2 + \frac{n\sigma\beta}{\sqrt{t}} $$

The maximal beta is then:
$$ \beta_{max,n} \sigma^2 = \mu + \frac{n\sigma}{\sqrt{t}} $$
$$ \beta_{max,n} = \frac{\mu}{\sigma^2} + \frac{n}{\sigma\sqrt{t}} $$

### Calculation of $\beta_{max,-1}$ for S&P 500 and QQQ

Using the formula:
$$
\beta_{max,n} = \frac{\mu}{\sigma^2} + \frac{n}{\sigma \sqrt{t}}
$$

For $n = -1$ and $t = 20$, we calculate $\beta_{max,-1}$ for S&P 500 and QQQ:

#### S&P 500:
- $\mu = 15.7\% = 0.157$
- $\sigma = 15\% = 0.15$
- $t = 20$

Substitute into the formula:
$$
\beta_{max,-1} = \frac{0.157}{(0.15)^2} + \frac{-1}{0.15 \sqrt{20}}
$$
$$
\beta_{max,-1} = \frac{0.157}{0.0225} - \frac{1}{0.67082}
$$
$$
\beta_{max,-1} \approx 6.978 - 1.49 \approx 5.488
$$

#### QQQ:
- $\mu = 22.0\% = 0.22$
- $\sigma = 22\% = 0.22$
- $t = 20$

Substitute into the formula:
$$
\beta_{max,-1} = \frac{0.22}{(0.22)^2} + \frac{-1}{0.22 \sqrt{20}}
$$
$$
\beta_{max,-1} = \frac{0.22}{0.0484} - \frac{1}{0.9832}
$$
$$
\beta_{max,-1} \approx 4.545 - 1.017 \approx 3.528
$$

These calculations provide the maximal leverage $\beta_{max,-1}$ for both indices.

### Calculation of $\mu_{max,n}$

To calculate $\mu_{max,n}$, we use the formula for $\mu_n$ with the maximal $\beta$:

$$
\mu_n = \beta \mu - \frac{1}{2} \beta^2 \sigma^2 + \frac{n \sigma \beta}{\sqrt{t}}
$$

Substituting $\beta_{max,n}$ into the equation:

$$
\beta_{max,n} = \frac{\mu}{\sigma^2} + \frac{n}{\sigma \sqrt{t}}
$$

We get:

$$
\mu_{max,n} = \left( \frac{\mu}{\sigma^2} + \frac{n}{\sigma \sqrt{t}} \right) \mu - \frac{1}{2} \left( \frac{\mu}{\sigma^2} + \frac{n}{\sigma \sqrt{t}} \right)^2 \sigma^2 + \frac{n \sigma}{\sqrt{t}} \left( \frac{\mu}{\sigma^2} + \frac{n}{\sigma \sqrt{t}} \right)
$$

After simplification, the expression for $\mu_{max,n}$ becomes:

$$
\mu_{max,n} = \frac{\mu^2}{2 \sigma^2} + \frac{n \mu}{\sqrt{t}} - \frac{n^2 \sigma^2}{2t}
$$

This represents the adjusted geometric growth rate $\mu_n$ with the maximal leverage $\beta$.

### Calculation of $\mu_{max,-1}$ for S&P 500 and QQQ

Using the formula:
$$
\mu_{max,n} = \frac{\mu^2}{2 \sigma^2} + \frac{n \mu}{\sqrt{t}} - \frac{n^2 \sigma^2}{2t}
$$

For $n = -1$ and $t = 20$, we calculate $\mu_{max,-1}$ for S&P 500 and QQQ:

#### S&P 500:
- $\mu = 15.7\% = 0.157$
- $\sigma = 15\% = 0.15$
- $t = 20$

Substitute into the formula:
$$
\mu_{max,-1} = \frac{(0.157)^2}{2 (0.15)^2} + \frac{-1 \cdot 0.157}{\sqrt{20}} - \frac{(-1)^2 (0.15)^2}{2 \cdot 20}
$$
$$
\mu_{max,-1} = \frac{0.02465}{2 \cdot 0.0225} - \frac{0.157}{4.472} - \frac{0.0225}{40}
$$
$$
\mu_{max,-1} = 0.5489 - 0.0351 - 0.0005625 \approx 0.5132 \text{ or } 51.32\%
$$

#### QQQ:
- $\mu = 22.0\% = 0.22$
- $\sigma = 22\% = 0.22$
- $t = 20$

Substitute into the formula:
$$
\mu_{max,-1} = \frac{(0.22)^2}{2 (0.22)^2} + \frac{-1 \cdot 0.22}{\sqrt{20}} - \frac{(-1)^2 (0.22)^2}{2 \cdot 20}
$$
$$
\mu_{max,-1} = \frac{0.0484}{2 \cdot 0.0484} - \frac{0.22}{4.472} - \frac{0.0484}{40}
$$
$$
\mu_{max,-1} = 0.5 - 0.0492 - 0.00121 \approx 0.4496 \text{ or } 44.96\%
$$

These calculations provide the adjusted geometric growth rates $\mu_{max,-1}$ for both indices with maximal leverage.

## Code

```python
import math

def calculate_mu_n(mu, sigma, n, t):
    """
    Calculate the adjusted geometric growth rate mu_n.
    
    Parameters:
    mu (float): GBM log-drift
    sigma (float): Annualized standard deviation
    n (int): Number of standard deviations from the median
    t (int): Time period in years
    
    Returns:
    float: Adjusted geometric growth rate mu_n
    """
    return mu - 0.5 * sigma**2 + (n * sigma / math.sqrt(t))

def calculate_beta_max(mu, sigma, n, t):
    """
    Calculate the maximal leverage beta_max.
    
    Parameters:
    mu (float): GBM log-drift
    sigma (float): Annualized standard deviation
    n (int): Number of standard deviations from the median
    t (int): Time period in years
    
    Returns:
    float: Maximal leverage beta_max
    """
    return (mu / sigma**2) + (n / (sigma * math.sqrt(t)))

def calculate_mu_n_max(mu, sigma, n, t):
    """
    Calculate the adjusted geometric growth rate mu_n with maximal leverage.
    
    Parameters:
    mu (float): GBM log-drift
    sigma (float): Annualized standard deviation
    n (int): Number of standard deviations from the median
    t (int): Time period in years
    
    Returns:
    float: Adjusted geometric growth rate mu_n with maximal leverage
    """
    term1 = (mu**2) / (2 * sigma**2)
    term2 = (n * mu) / math.sqrt(t)
    term3 = (n**2 * sigma**2) / (2 * t)
    return term1 + term2 - term3
```

```python
# Parameters for S&P 500
mu_sp500 = 0.157
sigma_sp500 = 0.15
n = -1
t = 20

# Parameters for QQQ
mu_qqq = 0.22
sigma_qqq = 0.22

# Calculate mu_n for S&P 500 and QQQ
mu_n_sp500 = calculate_mu_n(mu_sp500, sigma_sp500, n, t)
mu_n_qqq = calculate_mu_n(mu_qqq, sigma_qqq, n, t)

print(f"Adjusted geometric growth rate (mu_n) for S&P 500: {mu_n_sp500:.5f} or {mu_n_sp500 * 100:.3f}%")
print(f"Adjusted geometric growth rate (mu_n) for QQQ: {mu_n_qqq:.5f} or {mu_n_qqq * 100:.3f}%")

# Calculate beta_max for S&P 500 and QQQ
beta_max_sp500 = calculate_beta_max(mu_sp500, sigma_sp500, n, t)
beta_max_qqq = calculate_beta_max(mu_qqq, sigma_qqq, n, t)

print(f"Maximal leverage (beta_max) for S&P 500: {beta_max_sp500:.3f}")
print(f"Maximal leverage (beta_max) for QQQ: {beta_max_qqq:.3f}")

# Calculate mu_n_max for S&P 500 and QQQ
mu_n_max_sp500 = calculate_mu_n_max(mu_sp500, sigma_sp500, n, t)
mu_n_max_qqq = calculate_mu_n_max(mu_qqq, sigma_qqq, n, t)

print(f"Adjusted geometric growth rate with maximal leverage (mu_n_max) for S&P 500: {mu_n_max_sp500:.5f} or {mu_n_max_sp500 * 100:.3f}%")
print(f"Adjusted geometric growth rate with maximal leverage (mu_n_max) for QQQ: {mu_n_max_qqq:.5f} or {mu_n_max_qqq * 100:.3f}%")
```
