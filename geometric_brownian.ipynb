### Summary of Stock Value Using Geometric Brownian Motion

The Geometric Brownian Motion (GBM) is a mathematical model commonly used to describe the evolution of stock prices over time. The stochastic differential equation (SDE) for GBM is given by:

$$
dS_t = \mu S_t dt + \sigma S_t dW_t
$$

Where:
- $S_t$: Stock price at time $t$
- $\mu$: Drift coefficient (expected return)
- $\sigma$: Volatility of the stock
- $W_t$: Standard Brownian motion

The solution to this SDE can be derived using Itô's Lemma. Starting with the SDE:

$$
dS_t = \mu S_t dt + \sigma S_t dW_t
$$


The stochastic process $S_t$ evolves over time according to the stochastic differential equation $dS_t = \mu(S_t, t) dt + \sigma(S_t, t) dW_t$, where $\mu(S_t, t)$ represents the drift term and $\sigma(S_t, t) dW_t$ represents the stochastic term driven by Brownian motion. According to Ito's Lemma:

$$
df(S_t, t) = \left(\frac{\partial f}{\partial t} + \mu(S_t, t) \frac{\partial f}{\partial S} + \frac{1}{2} \frac{\partial^2 f}{\partial S^2} (\sigma(S_t, t))^2\right) dt + \sigma(S_t, t) \frac{\partial f}{\partial S} dW_t.
$$

Using Itô's Lemma, for $X_t = \ln(S_t)$, we calculate the derivatives:

1. First derivative with respect to $S_t$:
    $$
    \frac{\partial X_t}{\partial S_t} = \frac{\partial \ln(S_t)}{\partial S_t} = \frac{1}{S_t}
    $$

2. Second derivative with respect to $S_t$:
    $$
    \frac{\partial^2 X_t}{\partial S_t^2} = \frac{\partial}{\partial S_t} \left(\frac{1}{S_t}\right) = -\frac{1}{S_t^2}
    $$

3. Derivative with respect to $t$ vanishes since $X_t = \ln(S_t)$ does not explicitly depend on $t$:
    $$
    \frac{\partial X_t}{\partial t} = 0
    $$

Substituting these into Itô's Lemma:

$$
dX_t = \frac{\partial X_t}{\partial t} dt +  \mu S_t \frac{\partial X_t}{\partial S_t}dt + \frac{1}{2} \frac{\partial^2 X_t}{\partial S_t^2} (\sigma S_t)^2 dt + \sigma S_t \frac{\partial X_t}{\partial S_t} dW_t
$$


Therefore:

$$
dX_t = \mu dt + \frac{1}{2} \left(-\frac{1}{S_t^2}\right) (\sigma S_t)^2 dt + \sigma dW_t
$$

Simplify the second term:

$$
dX_t = \mu dt - \frac{1}{2} \sigma^2 dt + \sigma dW_t
$$

$$
dX_t = \left(\mu - \frac{\sigma^2}{2}\right) dt + \sigma dW_t
$$




Integrating both sides from $0$ to $t$:

$$
X_t = X_0 + \int_0^t \left(\mu - \frac{\sigma^2}{2}\right) dt + \int_0^t \sigma dW_t
$$

Since $X_t = \ln(S_t)$ and $X_0 = \ln(S_0)$:

$$
\ln(S_t) = \ln(S_0) + \left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t
$$

Exponentiating both sides gives the solution:

$$
S_t = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right)
$$


### Mean of $S_t$

The stock price $S_t$ follows the solution:

$$
S_t = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right)
$$

To calculate the mean of $S_t$, we use the expectation operator:

$$
\mathbb{E}[S_t] = \mathbb{E}\left[S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right)\right]
$$

Since everything except $W_t$ is a constant, it can be factored out:

$$
\mathbb{E}[S_t] = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t\right)\cdot\mathbb{E}\left[e^{\sigma W_t}\right]
$$


Calculating the $W_t$ part, we use the fact that if $W_t \sim \mathcal{N}(0, t)$, then $W_t$ has the probability density function:

$$
f_{W_t}(w) = \frac{1}{\sqrt{2 \pi t}} \exp\left(-\frac{w^2}{2t}\right), \quad w \in \mathbb{R}.
$$

The expectation of $\exp(\sigma W_t)$ is given by:

$$
\mathbb{E}\left[\exp\left(\sigma W_t\right)\right] = \int_{-\infty}^\infty \exp\left(\sigma w\right) f_{W_t}(w) \, dw.
$$

Substituting the PDF of $W_t$:

$$
\mathbb{E}\left[\exp\left(\sigma W_t\right)\right] = \int_{-\infty}^\infty \exp\left(\sigma w\right) \frac{1}{\sqrt{2 \pi t}} \exp\left(-\frac{w^2}{2t}\right) \, dw.
$$

Combine the exponential terms:

$$
\mathbb{E}\left[\exp\left(\sigma W_t\right)\right] = \frac{1}{\sqrt{2 \pi t}} \int_{-\infty}^\infty \exp\left(-\frac{w^2}{2t} + \sigma w\right) \, dw.
$$

Complete the square in the exponent $-\frac{w^2}{2t} + \sigma w$. Rewrite it as:

$$
-\frac{w^2}{2t} + \sigma w = -\frac{1}{2t}\left(w^2 - 2t\sigma w\right) = -\frac{1}{2t}\left((w - t\sigma)^2 - t^2\sigma^2\right).
$$

Simplify:

$$
-\frac{w^2}{2t} + \sigma w = -\frac{(w - t\sigma)^2}{2t} + \frac{t\sigma^2}{2}.
$$

Substitute this back into the integral:

$$
\mathbb{E}\left[\exp\left(\sigma W_t\right)\right] = \frac{1}{\sqrt{2 \pi t}} \int_{-\infty}^\infty \exp\left(-\frac{(w - t\sigma)^2}{2t}\right) \exp\left(\frac{t\sigma^2}{2}\right) \, dw.
$$

Factor out the constant $\exp\left(\frac{t\sigma^2}{2}\right)$:

$$
\mathbb{E}\left[\exp\left(\sigma W_t\right)\right] = \exp\left(\frac{t\sigma^2}{2}\right) \frac{1}{\sqrt{2 \pi t}} \int_{-\infty}^\infty \exp\left(-\frac{(w - t\sigma)^2}{2t}\right) \, dw.
$$

The remaining integral is the Gaussian integral of a normal distribution with mean $t\sigma$ and variance $t$, which evaluates to 1. Thus:

$$
\mathbb{E}\left[\exp\left(\sigma W_t\right)\right] = \exp\left(\frac{t\sigma^2}{2}\right).
$$




Substituting this back:

$$
\mathbb{E}[S_t] = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t\right) \exp\left(\frac{t \sigma^2}{2}\right)
$$

Simplify the exponents:

$$
\mathbb{E}[S_t] = S_0 \exp\left(\mu t\right)
$$

Thus, the mean of $S_t$ is:

$$
\mathbb{E}[S_t] = S_0 e^{\mu t}
$$


### Median of $S_t$

The stock price $S_t$ follows the solution:

$$
S_t = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t + \sigma W_t\right)
$$

The median of $S_t$ is the value $M_t$ such that $P(S_t \leq M_t) = 0.5$. Since $W_t \sim \mathcal{N}(0, t)$, the term $\sigma W_t$ has a median of 0. Substituting this into the equation for $S_t$, we get:

$$
M_t = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t\right)
$$

Thus, the median of $S_t$ is:

$$
\text{Median}(S_t) = S_0 \exp\left(\left(\mu - \frac{\sigma^2}{2}\right)t\right)
$$


### Converting to the PDF of Stock Value After Time $T$

At time $T$, the stock price $S_T$ follows a log-normal distribution. Taking the natural logarithm of $S_T$, we get:

$$
\ln(S_T) = \ln(S_0) + \left(\mu - \frac{\sigma^2}{2}\right)T + \sigma W_T
$$

Since $W_T \sim \mathcal{N}(0, T)$, the logarithm of the stock price $\ln(S_T)$ is normally distributed:

$$
\ln(S_T) \sim \mathcal{N}\left(\ln(S_0) + \left(\mu - \frac{\sigma^2}{2}\right)T, \sigma^2 T\right)
$$

Thus, the probability density function (PDF) of $S_T$ is:

$$
f_{S_T}(s) = \frac{1}{s \sigma \sqrt{2\pi T}} \exp\left(-\frac{\left(\ln(s) - \ln(S_0) - \left(\mu - \frac{\sigma^2}{2}\right)T\right)^2}{2\sigma^2 T}\right), \quad s > 0
$$

This is the PDF of a log-normal distribution.
